```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel(r"C:\Users\Avani Atul Patil\Downloads\netflix2.xlsx", sheet_name=0)

df.head()
```

```python
df.shape          # rows & columns
df.columns        # column names
df.info()         # data types
df.describe()     # numeric summary
```

```python
df.isnull().sum()
```

```python
df["director"] = df["director"].fillna("Unknown")
df["country"] = df["country"].fillna("Unknown")
```

```python
df["type"].value_counts()

```

```python
df["type"].value_counts().plot(kind="bar")
plt.title("Movies vs TV Shows")
plt.show()
```

```python
df["release_year"].value_counts().sort_index().plot()
plt.title("Content Released Over Years")
plt.show()
```

```python
sns.countplot(y="rating", data=df)
plt.title("Ratings Distribution")
plt.show()

```

```python
pd.pivot_table(df, values="release_year", index="type", aggfunc="mean")
```

```python


plt.scatter(df["release_year"], df.index)
plt.title("Number of Titles Over Years")
plt.xlabel("Release Year")
plt.ylabel("Index (Count of Titles)")
plt.show()
```

```python
import seaborn as sns

sns.boxplot(x="type", y="release_year", data=df)
plt.title("Release Year by Content Type")
plt.show()
```

```python
sns.histplot(df["release_year"], kde=True)
plt.title("Release Year Distribution")
plt.show()
```
